```python
# event loop:
    # greet customer
    # ask if they'd like to see the menu
        # iterate through objects (coffee, tea, pastries)
    # ask if they'd like to order 
    # receive item order
        # item validation
        # ask for quantity
    # ask if finished
    # display order
    # tabulate
    # 'have a nice day'

# menu objects
coffee = {
    'espresso' : 1.50,
    'americano' : 2.00,
    'caffe latte' : 2.25,
    'mocha' : 2.25,
    'cappucino' : 2.50,
    'macchiato' : 2.50
}
tea = {
    'white tea' : 1.50,
    'green tea' : 2.00,
    'oolong tea' : 2.25,
    'black tea' : 2.25,
    'earl grey' : 2.50,
    'matcha' : 2.50
}
pastries = {
    'croissant' : 5.50,
    'cinnamon roll' : 4.50,
    'banana bread' : 4.25,
    'pumpkin bread' : 3.75,
    'sticky bun' :  4.25,
    'apple turnover' : 3.75
}

#all menus dict
all_menus = {}
all_menus['coffee'] = coffee
all_menus['tea'] = tea
all_menus['pastries'] = pastries


# all items dict to make ordering and tabulation easier
all_items = {}
all_items.update(coffee)
all_items.update(tea)
all_items.update(pastries)


#  display menus function  
def display_menu(obj) :
    for menu in obj.keys():
        print(f"{menu.title()}")
        for key, value in obj[menu].items(): # <- hell yeah v proud of this line
            print(f"{key.title()} : {value:.2f}")


#get order
def add_order(item) :
    foo = []
    quant = int(input(f'How many {item.title()}s would you like?')) 
    for i in range(quant) :
        foo.append( item )
    return foo
        
        

def place_order() :
    ordering = ''
    tab = []
    while ordering != 'n':
        item_order = input('What would you like? :')
        item_order = item_order.lower()
        if item_order in all_items :
            temp = add_order(item_order)
            # flattening list
            for i in temp :
                tab.append(i)
        else :
            print("Sorry, that's not on the menu :(") 
            
        ordering = input('Would you like anything else? (y/n): ')
        ordering = ordering.lower()
    return tab

def print_receipt(order):
    total = 0
    sub_total = 0
    tax = .06
    print('\nYour order is: \n')
    
    for key, values in order.items() :
        print(f'{values} * {key.title()} = {all_items[key]*values:.2f}')
        sub_total += all_items[key]*values
    print(f'\nSub-Total: \t{sub_total:.2f} \nTax: \t\t{sub_total*tax:.2f}\nTotal: \t\t{sub_total+(sub_total*tax):.2f}\n')

def cafe_loop() :
    print('~ Welcome to The Cafe! ~')
    
    # see menu
    ans = input('Would You like to see the menu? (y/n): ')
    ans = ans.lower()
    if ans != 'n' :
        display_menu(all_menus)
        
    # place order function
    receipt = []
    receipt = place_order()
    
    # count items
    order_dict = {i:receipt.count(i) for i in receipt}
    # print receipt
    print_receipt(order_dict)

    print('Have a nice day!')



#execute program
cafe_loop()
    
```
